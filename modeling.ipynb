## Filtering unused columns

```python
cols = [
    # movie name
    'movieNm',
    # features
    'audiAcc',
    'scrnCnt',
    'showCnt',
    'genres',
    'director_career_length',
    'avg score',
    'avg aud',
    'movie_length',
    'mean_actor_nummovie',
    'max_actor_nummovie',
    'main_actor_nummovie',
    # 'PM10_mean',
    # 'PM10_max',
    # 'PM10_bad_days',
    # 'PM25_mean',
    # 'PM25_max',
    # 'PM25_bad_days',
    # 
    'openDt',
    'directors',
    'actors',
    'company_names',
    'director_prev_movie',
    'director_tot_filmo',
]
df = df[cols]
df
```

```python
df = df[(df['directors'].map(ast.literal_eval).map(len) > 0) & (df['actors'].map(ast.literal_eval).map(len) > 0)]
df
```

## Make genre columns

```python
for genre in df['genres'].map(ast.literal_eval).explode().value_counts().index:
    df[f'genre_{genre}'] = df['genres'].map(ast.literal_eval).map(lambda genres: genre in genres)
df
```

## Add # of aud features

### Director

```python
aud_by_dir = pd.read_csv('audmaxmean_bydir.csv')
aud_by_dir
```

```python
movie_directors = df[['movieNm', 'directors']]
movie_directors['directors'] = movie_directors['directors'].map(ast.literal_eval)
movie_directors.explode('directors')
```

```python
df_merged = movie_directors.explode('directors').merge(aud_by_dir, how = 'left', left_on = 'directors', right_on = 'peopleNm')
df_merged = df_merged.dropna(subset=['peopleNm'])
df_merged = df_merged.groupby('movieNm').agg({'max_audiAcc': 'max', 'mean_audiAcc': 'mean'})
df_merged
```

```python
df = pd.merge(df, df_merged.add_suffix('_director'), how = 'left', on = 'movieNm', ).dropna(subset=['max_audiAcc_director'])
df
```

### Company

```python
aud_by_comp = pd.read_csv('audmaxmean_bycomp.csv')
aud_by_comp
```

```python
movie_companys = df[['movieNm', 'company_names']]
movie_companys['company_names'] = movie_companys['company_names'].map(ast.literal_eval)
movie_companys.explode('company_names')
```

```python
df_merged = movie_companys.explode('company_names').merge(aud_by_comp, how = 'left', left_on = 'company_names', right_on = 'companyNm')
df_merged = df_merged.dropna(subset=['companyNm'])
df_merged = df_merged.groupby('movieNm').agg({'max_audiAcc': 'max', 'mean_audiAcc': 'mean'})
df_merged
```

```python
df = pd.merge(df, df_merged.add_suffix('_company'), how = 'left', on = 'movieNm', ).dropna(subset=['max_audiAcc_company'])
df
```

```python
df.to_csv('data_merged.csv', index = False)
```
