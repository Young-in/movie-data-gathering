```python
import pandas as pd
```

```python
movie_info = pd.read_csv("movies.csv")
peop_info = pd.read_csv("people_info.csv")
print(movie_info)
```

```python
df = pd.DataFrame(movie_info)
print(df)
```

```python
l = len(df["actors"])

cd_list=[]
act_name_list=[]

```

```python
actor_dict = dict()
actor_filmos = dict()

for i in range(l):
  a= df["actors"][i].split("'")
  name = df["movieNm"][i]
  doub_check = set()
  x = len(a)
  for j in range(x):
    if a[j] == "peopleNm":
      act_name = a[j+2]
      ind_list =  peop_info[peop_info["peopleNm"]==act_name].index.tolist()
      for t in ind_list:
        if peop_info.loc[t]["repRoleNm"]=="배우":
          r = peop_info.loc[t]["filmos"].split("'")
          istrue=0
          for q in range(len(r)):
            if r[q] == name:
              istrue = 1
              break
          if istrue:
            if act_name in doub_check:
              continue
            doub_check.add(act_name)
            cd = peop_info.loc[t]["peopleCd"]
            if cd in actor_dict:
              actor_dict[cd] += 1
              actor_filmos[cd].append(name)
            else:
              actor_dict[cd] = 1
              actor_filmos[cd]=[]
              actor_filmos[cd].append(name)
              cd_list.append(cd)
              act_name_list.append(act_name)
      
```

```python
act_name_list
```

```python
ll = len(cd_list)
act_num = [None]*ll
act_filmos=[None]*ll


for i in range(ll):
  act_num[i]=actor_dict[cd_list[i]]
  act_filmos[i]=actor_filmos[cd_list[i]]

```

```python
df_final = pd.DataFrame({
    "Cd" : cd_list,
    "name" : act_name_list,
    "num_Movie" : act_num,
    "filmos" : act_filmos
})

df_final
```

```python
df_final.to_csv("actor_NumMovies.csv")
```
